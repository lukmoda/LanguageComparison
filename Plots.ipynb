```python
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')


n_groups = 3
vals_fort = fort.iloc[[10, 17, 29],-1].values
vals_py = py.iloc[[10, 17, 29],-1].values
vals_jl = jl.iloc[[10, 17, 29],-1].values
vals_r = r.iloc[[10, 17, 29],-1].values

# create plot
fig, ax = plt.subplots(figsize=(12, 8))
index = np.arange(n_groups)
bar_width = 0.1
opacity = 0.8

rects1 = plt.bar(index, vals_jl, bar_width,
alpha=opacity,
color='b',
label='Julia')

rects2 = plt.bar(index+bar_width, vals_fort, bar_width,
alpha=opacity,
color='r',
label='Fortran')

rects3 = plt.bar(index+bar_width*2, vals_py, bar_width,
alpha=opacity,
color='g',
label='Python')

rects4 = plt.bar(index+bar_width*3, vals_r, bar_width,
alpha=opacity,
color='cyan',
label='R')

plt.xlabel('Number of Needles')
plt.ylabel('Time of Execution')
plt.title('Time of Execution by Language')
plt.xticks(index + bar_width, ('10k', '100M', '2G'))
plt.legend()

plt.tight_layout()
plt.yscale('log')
autolabel(rects1, "left")
autolabel(rects2, "left")
autolabel(rects3, "left")
autolabel(rects4, "left")
plt.show()
```

### Random Forest (Train)

#### Python
